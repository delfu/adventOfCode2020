```python
from copy import deepcopy
def transition(pocket):
    ret_pocket = deepcopy(pocket)
    acc = {}
    dim_min = [0, 0, 0]
    dim_max = [0, 0, 0]
    for key in pocket.keys():
        x, y, z = key.split(",")
        x = int(x)
        y = int(y)
        z = int(z)
        dim_min = [min(x, dim_min[0]), min(y, dim_min[1]), min(z, dim_min[2])]
        dim_max = [max(x+1, dim_max[0]), max(y + 1, dim_max[1]), max(z + 1, dim_max[2])]
        for i in range(-1, 2):
            for j in range(-1, 2):
                for k in range(-1, 2):
                    if i == j == k == 0:
                        continue
                    if key in pocket and pocket[key] == "#":
                        neighbor_key = str(x+i) + "," + str(y+j) + "," + str(z+k)
                        acc[neighbor_key] = 1 if neighbor_key not in acc else acc[neighbor_key] + 1
    dim_min = [axis - 1 for axis in dim_min]
    dim_max = [axis + 1 for axis in dim_max]
    for x in range(dim_min[0], dim_max[0]):
        for y in range(dim_min[1], dim_max[1]):
            for z in range(dim_min[2], dim_max[2]):
                key = str(x) + "," + str(y) + "," + str(z)
                if key in pocket and pocket[key] == "#":
                    if key in acc and (acc[key] == 2 or acc[key] == 3):
                        ret_pocket[key] = "#"
                    else: 
                        ret_pocket[key] = "."
                else:
                    if key in acc and acc[key] == 3:
                        ret_pocket[key] = "#"
                    else: 
                        ret_pocket[key] = "."
    return ret_pocket, dim_min, dim_max

def print_pocket(pocket, dim_min, dim_max):
    active = 0
    for z in range(dim_min[2], dim_max[2]):
        print("z="+str(z))
        for y in range(dim_min[1], dim_max[1]):
            line = ""
            for x in range(dim_min[0], dim_max[0]):
                key = str(x) + "," + str(y) + "," + str(z)
                if key not in pocket:
                    line += "."
                else:
                    line += pocket[key]
                    active += 1 if pocket[key] == "#" else 0
            print(line)
        print("")
    return active
                    

def solution1():
    initial = [case for case in util.input_reader('day17.input', 1)]
    pocket = {}
    for r, row in enumerate(initial):
        for c, cell in enumerate(row):
            if cell == "#": 
                pocket[str(c) + "," + str(r) + "," + str(0)] = "#"
    # print(pocket)
    for cycle in range(6):
        pocket, dim_min, dim_max = transition(pocket)
        # print_pocket(pocket, dim_min, dim_max)
        # print(dim_min, dim_max)
    return print_pocket(pocket, dim_min, dim_max)
    # return pocket
solution1()
```
